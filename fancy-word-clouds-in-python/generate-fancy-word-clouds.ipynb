```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image
```

```python
text = "Hello World"

wc = WordCloud().generate(text)
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
text = open('theory behind earliest invasion', 'r').read()

wc = WordCloud().generate(text)
plt.imshow(wc)
plt.axis("off")
plt.savefig('history.png')
plt.show()
```

```python
text = open('theory behind earliest invasion', 'r').read()

wc = WordCloud(stopwords=STOPWORDS).generate(text)
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
text = open("cover letter.txt", 'r').read()

wc = WordCloud(background_color="white").generate(text)
plt.imshow(wc)
plt.axis("off")
plt.savefig('cover_letter2.png')
plt.show()
```

```python
text = open("cover letter.txt", 'r').read()

chrome_mask = np.array(PIL.Image.open("chrome-logo.jpg"))

colormap = ImageColorGenerator(chrome_mask)

wc = WordCloud(mask=chrome_mask, 
               background_color="white",
              contour_color="black",
              contour_width=2,
              min_font_size=3,
              max_words=50).generate(text)

wc.recolor(color_func = colormap)

plt.figure(figsize=(10,8))
plt.imshow(wc)
plt.axis("off")
# plt.savefig('cover_letter3.png')
plt.show()
```

```python
text = open("Lover's poem to his love", 'r').read()

chrome_mask = np.array(PIL.Image.open("chrome-logo.jpg"))

colormap = ImageColorGenerator(chrome_mask)

wc = WordCloud(mask=chrome_mask, 
               background_color="white",
              contour_color="brown",
              contour_width=2,
              min_font_size=3,
              max_words=100).generate(text)

wc.recolor(color_func = colormap)

plt.figure(figsize=(10,8))
plt.imshow(wc)
plt.axis("off")
plt.savefig('god-love.png')
plt.show()
```
